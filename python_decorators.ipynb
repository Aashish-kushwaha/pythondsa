## Decorators
a decorators is a essentially a function tat takes another functions as an arguments and returns a new function with enhanced functionality 

```python
def decorator(func):
    def wrapper():
        print("before calling the function")
        func()
        print("after calling the function")
    return wrapper

@decorator
def greet():
    print("hello worls")

greet()
```

## decorators with parameters:
decorator often need to work wih functions that have arguments. we use *args and **kargs so our wrapper can accept any number of arguments.

```python
def decorator_name(func):
    def wrapper(*args):
        print("before excution")
        r=func(*args)
        print("after excution")
        return r
        
    return wrapper

@decorator_name
def addition(a,b):
    return a+b

addition(10,20)
```

## function as a first-class objects
in python function are treated as first class citizens , meaning they can be treated like any other object(int, string, list). this allows function to be assigned to variables, passed as arguments, return from other function

```python
# Assigning a function to a variable
def greet(n):
    return f"Hello, {n}!"
say_hi = greet  # Assign the greet function to say_hi
print(say_hi("Alice")) 

# Passing a function as an argument
def apply(f, v):
    return f(v)
res = apply(say_hi, "Bob")
print(res) 

# Returning a function from another function
def make_mult(f):
    def mult(x):
        return x * f
    return mult
dbl = make_mult(2)
print(dbl(5))
```

## high order function
these are the function that take one or more function as arguments, return a function as a resut or do both, essentially, a hogher order function operates on other function.

```python
# A higher-order function that takes another function as an argument
def fun(f, x):
    return f(x)

# A simple function to pass
def square(x):
    return x * x
res = fun(square, 5) # Using apply_function to apply the square function
print(res)
```
