# API handling

```python
import requests
```

```python
url="https://api.freeapi.app/api/v1/public/randomusers/user/random"
```

```python
response=requests.get(url)
print(response)
print(type(response))
```

```python
df=response.json()
print(df)
print(type(df))
```

```python
print(df["success"])
```

```python
for i in df:
    print(i)
    print(type(i))
```

```python
if df["success"] and "data" in df:
    user_data=df["data"]
    print(user_data)
    # print(user_data["login"]["username"])
```
