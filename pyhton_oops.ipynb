### points to remeber
* classes are created by a keyword ```class```
* attributes are variables that belongs to a class
* attributes are always public and can be accessed using a dot (.)


````python
#example of class
class dogs:
    species="canine"

    def __init__(self,Name,age):
        self.name=Name
        self.age=age
````

## Self parameter
it is a reference to the current instance of the class. it allows us to access the attribtes and methods of the objects

````python
class Dog:
    species = "Canine"  # Class attribute

    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age  # Instance attribute

dog1 = Dog("Buddy", 3)  # Create an instance of Dog
dog2 = Dog("Charlie", 5)  # Create another instance of Dog

print(dog1.name, dog1.age, dog1.species)  # Access instance and class attributes
print(dog2.name, dog2.age, dog2.species)  # Access instance and class attributes
print(Dog.species)  # Access class attribute directly
````

### types of variables in a class
class variables: these are shared variables that are shared across all instances of a class. it is defined at the class level, outside any methods. all objects of the class share the same value for a class variable unless explicitly overridden in an object

````python
class dogs:
    species="canine" # class variables

    def __init__(self,name,age):
        self.name=name
        self.age=age
    
````

## inheritance
inheritance allows a class(child) to acquire properties and methods of another class(parents). it supports hierarchical and promotes code reuse.
### types of inheritance
1. single inheritance: a child class inherits from a single parent class
2. multiple inheritance: a child class inherits from a parent class, which in turn inherits from another class.
3. Hierarchical inheritance: multiple child classes inherit from a single parent class
4. hybrid inheritance: a combination of two or more types of inheritance



````python
# single inheritance
class Animal:
    def sound(self):
        return "some sound"

class Dog(Animal):
    def bark(self):
        return "woof"

d=Dog()
print(d.sound())
print(d.bark())


````

````python
# multiple inheritance
class Father:
    def father_skill(self):
        return "driving"

class Mother:
    def mother_skill(self):
        return "cooking"

class child(Father,Mother):
    def child_skill(self):
        return "chess"

c=child()
print(c.father_skill())
print(c.mother_skill())
print(c.child_skill())
````

````python
# multilevel inheritance
class Grand_father:
    def gf_property(self):
        return "land" 

class Father(Grand_father):
    def f_property(self):
        return "house"
class Son(Father):
    def s_property(self):
        return "laptop"
s=Son()
print(s.gf_property())
print(s.f_property())
print(s.s_property())
````

````python
#hierarchical inheritance
class Parent:
    def parent_method(self):
        return "Parent property"

class Child1(Parent):
    def c1_method(self):
        return "Child1 property"

class Child2(Parent):
    def c2_method(self):
        return "Child2 property"

c1 = Child1()
c2 = Child2()
print(c1.parent_method(), c1.c1_method())
print(c2.parent_method(), c2.c2_method())
````

````python
#hybrid inheritance

# Base Parent
class A:
    def method_a(self):
        return "A"

# B inherits from A (multilevel)
class B(A):
    def method_b(self):
        return "B"

# C also inherits from A (hierarchical)
class C(A):
    def method_c(self):
        return "C"

# D inherits from B and C (multiple)
class D(B, C):
    def method_d(self):
        return "D"

d = D()
print(d.method_a())  # from A
print(d.method_b())  # from B
print(d.method_c())  # from C
print(d.method_d())  # own method
````

## polymorphism
polymorphism means same operation ,different behaivour" it allows functions or methods with same name to work differently depending on the type of object they are acting upon
types of polymorphism
1. compile time polymorphism
2. runtime polymorphism

### complie time polymorphism
ths type of polymorphism is determined during the compilation of the program . it allows methods or operator with the same name to behave differently based on their input parameters or usage.

in python:
* true compile-time polymorphism is not supported
* python mimics it using default arguments or  ```*args/**krgs```


````python
class Calculator:
    def add(self, *args):
        return sum(args)
c=Calculator()
print(c.add(1,2))
print(c.add(1,2,3))
print(c.add(1,2,3,4))
````

### run-time polymorphism
Run-time polymorphism is determined during the execution of the program.
1. method overriding: a subclass redefines a method from its parents class
2. duck tying: if an objects implements the required method , it works regardless of its type
3. operator overloading: special methods ```(__add__, __sub__)``` redefine how operators behave for user-defined objects
