## python closure

closures in python allows functions to remember values fron the environment in which it was created even if the environment is no longer exists

### how closures are created
* a function is defined inside another function( nested function)
* the inner function references the variables from the outer functions
* the outer function returns the inner function

```python
# nested function
def outer():
    def inner():
        print("hello")
    inner()
outer()
```

```python
# inner function uses the variable of outer function
def outer(a):
    b=10
    def inner():
        print(a*b)
    inner()

outer(2)
```

```python
#outer function return the inner function
def outer():
    x=9
    def inner():
        print("hello")
    return inner

c=outer()
print(type(c))
c()
```

### why closure are usefulls:
closures are usefull because they allow to maintain the state without using 
* global variables
* classes
* external storage

Benefits: \
1️⃣ Data hiding / Encapsulation

Closures let you “hide” data inside a function.

2️⃣ Creating function factories

You can generate customized functions with preset parameters.

3️⃣ Maintaining state

Closures keep track of values across multiple calls.

4️⃣ Used in decorators

Decorators rely heavily on closures to wrap functions.

```python
#creating function factories
def multiplier(n):
    def multiply(x):
        return n*x
    return multiply

a=multiplier(10)
a(9)
```

```python
#maintaining the state without classes
def make_counter():
    count=0
    def counter():
        nonlocal count  # The nonlocal keyword in Python is used within nested functions to declare that a variable refers to a previously bound variable in the nearest enclosing but non-global scope. This allows the inner function to modify the value of a variable defined in the outer function, which is not the default behavior.  
        count+=1

        return count
    return counter

c=make_counter()
print(c())

print(c())
```
